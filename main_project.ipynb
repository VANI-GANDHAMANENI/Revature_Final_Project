```python
import yfinance as yf

# Define the stock ticker symbol (Change this to any stock symbol you want)
ticker_symbol = "AAPL"  # Example: Apple Inc.

# Fetch data from Yahoo Finance
company = yf.Ticker(ticker_symbol)

# Fetch historical stock prices (Last 1 month)
stock_prices = company.history(period="1mo")
income_statement = company.financials  # Income Statement
balance_sheet = company.balance_sheet  # Balance Sheet
cash_flow = company.cashflow  # Cash Flow Statement

# Print sample outputs
print("Stock Prices:\n", stock_prices.head())
print("\nIncome Statement:\n", income_statement)
print("\nBalance Sheet:\n", balance_sheet)
print("\nCash Flow Statement:\n", cash_flow)

# Optional: Save data to CSV files
stock_prices.to_csv("stock_prices.csv")
income_statement.to_csv("income_statement.csv")
balance_sheet.to_csv("balance_sheet.csv")
cash_flow.to_csv("cash_flow.csv")

print("\nData successfully fetched and saved as CSV files!")
```

Drop rows with missing values

```python
financial_stmts.dropna(inplace=True)
dataincome_statements.dropna(inplace=True)
balance_sheets.dropna(inplace=True)
cash_flows.dropna(inplace=True)
companies.dropna(inplace=True)
```

```python
print("Financial Statements Columns:", financial_stmts.columns.tolist())
print("Balance Sheets Columns:", balance_sheets.columns.tolist())
print("Cash Flow Columns:", cash_flows.columns.tolist())
print("Income Statements Columns:", dataincome_statements.columns.tolist())
```

```python
dataincome_statements['Profit_Margin'] = dataincome_statements['Net Income'] / dataincome_statements['Total Revenue']
```

```python
dataincome_statements.columns = dataincome_statements.columns.str.strip().str.lower()
balance_sheets.columns = balance_sheets.columns.str.strip().str.lower()
cash_flows.columns = cash_flows.columns.str.strip().str.lower()
financial_stmts.columns = financial_stmts.columns.str.strip().str.lower()
```

```python
financial_stmts.rename(columns={"year": "year", "company ": "company"}, inplace=True)  # Fix space issue
balance_sheets.rename(columns={"unnamed: 0": "year"}, inplace=True)
cash_flows.rename(columns={"unnamed: 0": "year"}, inplace=True)
dataincome_statements.rename(columns={"unnamed: 0": "year"}, inplace=True)
```

```python
print("Financial Statements Columns:", financial_stmts.columns.tolist())

# Check if "year" exists, if not rename the correct column
if "year" not in financial_stmts.columns and "Year" in financial_stmts.columns:
    financial_stmts.rename(columns={"Year": "year"}, inplace=True)

# Convert "year" to integer
financial_stmts["year"] = financial_stmts["year"].astype(int)
balance_sheets["year"] = balance_sheets["year"].astype(str).str.extract("(\d+)").astype(float).astype("Int64")
cash_flows["year"] = cash_flows["year"].astype(str).str.extract("(\d+)").astype(float).astype("Int64")
dataincome_statements["year"] = dataincome_statements["year"].astype(str).str.extract("(\d+)").astype(float).astype("Int64")

# Merge datasets on "year"
df_merged = financial_stmts.merge(balance_sheets, on="year", how="left")
df_merged = df_merged.merge(cash_flows, on="year", how="left")
df_merged = df_merged.merge(dataincome_statements, on="year", how="left")
df_merged.fillna(0, inplace=True)

# Save the merged dataset
df_merged.to_csv(r'C:\Users\My Pc\.kaggle\data\final_financial_data.csv', index=True)

print("✅ Data Merging Successful!")
```

```python
import streamlit as st
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt

def fetch_financial_data(ticker):
    company = yf.Ticker(ticker)
    return {
        "Stock Prices": company.history(period="1mo"),
        "Income Statement": company.financials,
        "Balance Sheet": company.balance_sheet,
        "Cash Flow Statement": company.cashflow,
    }

def main():
    st.title("Financial Statement Generator")
    
    # Company selection
    ticker = st.text_input("Enter Stock Ticker Symbol (e.g., AAPL, TSLA, MSFT)", "AAPL")
    if st.button("Generate Report"):
        data = fetch_financial_data(ticker)
        
        # Display financial statements with drill-down
        for key, df in data.items():
            st.subheader(key)
            st.write(df.head())
            
            # Visualization for trends
            if key == "Stock Prices":
                st.line_chart(df['Close'])
            
        # Export functionality
        st.download_button("Download Income Statement", data["Income Statement"].to_csv(), "income_statement.csv")
        st.download_button("Download Balance Sheet", data["Balance Sheet"].to_csv(), "balance_sheet.csv")
        st.download_button("Download Cash Flow", data["Cash Flow Statement"].to_csv(), "cash_flow.csv")

if __name__ == "__main__":
    main()
```
